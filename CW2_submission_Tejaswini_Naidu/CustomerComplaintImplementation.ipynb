#### Balancing the Dataset is required

In our analysis, we observed that two product categories, **Credit reporting, repair or reports** and **Debt collection**, have significantly higher complaint counts compared to the remaining three categories: **Credit or prepaid card**, **Checking or savings account**, and **Mortgage**. To achieve a more balanced dataset and avoid potential biases in our analysis, we decided to undersample the **Credit reporting** and **Debt collection** categories. 

##### 1. Undersampling **Credit Reporting**:

The **Credit reporting** category had an exceptionally high complaint count, approximately 10.98 times that of the "Mortgage" category. By undersampling this category, we aim to reduce its contribution to the analysis, making it more comparable to the other three categories. While it won't be a perfect match, this approach allows us to mitigate the skewness caused by the overwhelming number of **Credit reporting** complaints.

##### 2. Undersampling **Debt Collection**:

Similar to **Credit reporting**, the **Debt collection** category had a significantly higher complaint count, approximately 2.33 times that of "Mortgage." Undersampling this category brings it closer in terms of complaint counts to the remaining three categories, reducing the potential bias it may introduce into the analysis.

##### 3. Keeping the Other Three Categories As Identicals:

Conversely, we decided to keep the complaint counts for the other three categories (**Credit or prepaid card**, **Checking or savings account**, and **Mortgage**) unchanged. These categories had relatively fewer complaints compared to **Credit reporting** and **Debt collection**, but the differences were not as extreme. Maintaining the original counts for these categories allows us to analyze them in their natural distribution, as the differences in counts are not as significant as the two aforementioned categories.

This approach aims to strike a balance between fairness in the analysis and the practicality of handling imbalanced data. It enables us to explore insights from a dataset that better represents the distribution of complaints across the selected product categories.

Certainly, here's the explanation in Jupyter Notebook markdown format:

#### How to balance the dataset now with downsampling in Code

1. **Calculate Average Count:** Compute the average count of complaints within the three selected categories (which shows lower counts in top 5). This average count will serve as the target count for undersampling the other two high count categories.

2. **Undersample Specific Categories:** For each category you want to balance (e.g., 'Credit reporting' and 'Debt collection'), select all the complaints within that category. Undersample the category by randomly selecting a fraction of the complaints to match the average count calculated earlier.

3. **Combine Undersampled Data:** Merge the undersampled complaints from each category into a new DataFrame.

4. **Combine with Unchanged Categories:** Combine the undersampled top products with the unchanged complaints from the initial subset.

````python
def balance_dataset(df):
    """
    Balances the dataset by undersampling 'Credit reporting' and 'Debt collection' to match the average count 
    of 'Credit card or prepaid card', 'Checking or savings account', and 'Mortgage'.

    Args:
        df (pd.DataFrame): The input DataFrame containing the dataset.

    Returns:
        pd.DataFrame: The balanced dataset.
    """

    # Filter the DataFrame for the top 5 products
    temp_df = df[df['Product'].isin(['Credit card or prepaid card', 'Checking or savings account', 'Mortgage'])]

    # Calculate the average count of the above three categories
    average_count = temp_df.groupby('Product').size().mean()

    # Determine the desired undersampling ratio for Credit reporting and Debt collection
    # 1. Undersample Credit reporting
    df_credit_reporting = df[df['Product'] == 'Credit reporting, credit repair services, or other personal consumer reports']
    undersampled_credit_reporting = df_credit_reporting.sample(frac=average_count / len(df_credit_reporting), 
                                                           random_state=42)

    # 2. Undersample Debt collection
    df_debt_collection = df[df['Product'] == 'Debt collection']
    undersampled_debt_collection = df_debt_collection.sample(frac=average_count / len(df_debt_collection), 
                                                         random_state=42)

    # Combine the undersampled data
    undersampled_df_top = pd.concat([undersampled_credit_reporting, undersampled_debt_collection])

    # Combine the undersampled top products with the unchanged other categories
    new_df = pd.concat([undersampled_df_top, temp_df])

    # Plot Products distribution
    explore_product(new_df)

    return new_df

balanced_df = balance_dataset(filtered_df)
````

````python
import plotly.express as px
from plotly.subplots import make_subplots

# Create a bar plot function
def plot_product_distribution(df, title, yaxis_title):
    data = df['Product'].value_counts()
    data = data.sort_values(ascending=False)

    fig = px.bar(data, x=data.index, y=data.values, color=data.index,
                 labels={'x': 'Product', 'y': 'Count', 'color': 'Product'},
                 title=title)

    fig.update_traces(texttemplate='%{y}', textposition='outside',
                      textfont_size=8)

    fig.add_annotation(x=0.5, y=1.1,
                       text="Total Complaints: " + str(data.sum()),
                       showarrow=False,
                       font=dict(size=12, color="black"),
                       xref="paper",
                       yref="paper")

    fig.update_xaxes(showticklabels=False)
    # replace 'index' with 'Product' on x-axis
    fig.update_layout(xaxis_title="Product", yaxis_title=yaxis_title)

    return fig

# Create subplots
fig = make_subplots(rows=1, cols=2, shared_yaxes=True, horizontal_spacing=0.1)

# Plot Product distribution for filtered_df
fig_filtered = plot_product_distribution(filtered_df, 'Product By Complaint Volume (Original)', 
                                          'Count (Original)')
fig.update_layout(showlegend=False)
for trace in fig_filtered.data:
    fig.add_trace(trace, row=1, col=1)

# Plot Product distribution for balanced_df
fig_balanced = plot_product_distribution(balanced_df, 'Product By Complaint Volume (Balanced)',
                                         'Count (Balanced)')
fig.update_layout(showlegend=False)
for trace in fig_balanced.data:
    fig.add_trace(trace, row=1, col=2)

# Update layout
fig.update_layout(height=500, width=950, 
                  title_text="Product Distribution Comparison (Before and After Balancing/Downsampling)")
fig.show()
````

#### 3.4.1.b Data Preprocessing and Feature Transformation
##### 1. Text Data Preprocessing Steps

To illustrate these tasks, we provide an example below, demonstrating how a sample document will be processed at each step:

| Task                                           | Description                                                                                                                                                                  | Example                |
|-----------------------------------------------|------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|------------------------|
| Original Document (doc)                        | Original unaltered document text.                                                                                                                                                  | "I had a terrible experience with their customer service department." |
| Text Normalization (doc_lower)                 | Convert all letters to lowercase to ensure uniformity and consistency across all documents.                                                                               | "i had a terrible experience with their customer service department." |
| Special Character Removal and Tokenization (doc_tokens) | Eliminate special characters and tokenize the remaining content into individual terms (words).            | ["i", "had", "a", "terrible", "experience", "with", "their", "customer", "service", "department"] |
| Stop Word Filtering (doc_filtered)             | Remove common stop words to focus on meaningful words and reduce feature noise.       | ["terrible", "experience", "customer", "service", "department"] |
| Stemming - `Porter Stemmer` (doc_stemmed)                         | Reduce words to their root form to treat different word forms as a single term.                          | ["terribl", "experienc", "custom", "servic", "depart"] |
| Token Generation and Filtering (doc_filtered_tokens)  | Process the remaining terms after stemming to generate tokens for the term-document matrix.                                                  | ["terribl", "experienc", "custom", "servic", "depart"] |

`Purpose of Stemming in Text Analysis:` The purpose of stemming in text analysis is to simplify and standardize text by reducing words to their base or root form. Stemming helps in handling variations of words more efficiently. For example, words like "protect," "protection," and "protecting" are stemmed to their base form, which is "protect." This step is crucial for improving text classification analysis by focusing on the core meaning of words rather than considering multiple variations of the same word.

`Benefits of Stemming:`
1. **Reducing Variations:** Stemming reduces words to their base form, eliminating unnecessary variations. This simplification makes it easier to analyze text data.

2. **Resource Efficiency:** Text analysis algorithms can be resource-intensive when dealing with numerous word forms. Stemming helps in conserving computational resources by treating variations as the same word.

3. **Avoiding Confusion:** Stemming reduces ambiguity and confusion in text mining algorithms. It ensures that different forms of the same word are treated as identical, leading to more accurate and consistent results.

Here, `The Porter Stemmer` is a widely used algorithm for stemming, which is the process of reducing words to their root or base form. It was developed by Martin F. Porter in 1980 and is one of the most well-known stemming algorithms. The Porter Stemmer algorithm works by removing common suffixes from words to obtain their root form. This can help in treating different forms of a word as the same, which is useful in natural language processing tasks like text analysis and information retrieval.

For example, the Porter Stemmer would convert words like "jumps," "jumping," and "jumped" to their common root form "jump."

More Information: To delve deeper into the details of the Porter Stemmer algorithm, you can refer to the [original paper by Martin F. Porter](https://tartarus.org/martin/PorterStemmer/def.txt) where he explains the algorithm and provides examples of its application.


````python
import re
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords

def text_transformation(original_doc):
    """
    Apply text transformation tasks to the original document.

    Args:
        original_doc (str): The original unaltered document text.

    Returns:
        list of str: A list of transformed versions of the document.
    """
    
    # Task 1: Convert to lowercase
    doc_lower = original_doc.lower()
    
    # Task 2: Special Character Removal and Tokenization
    doc_tokens = re.findall(r'\b\w+\b', doc_lower)
    
    # Task 3: Stop Word Filtering
    stop_words = set(stopwords.words("english"))
    doc_filtered = [word for word in doc_tokens if word not in stop_words]
    
    # Task 4: Stemming
    stemmer = PorterStemmer()
    doc_stemmed = [stemmer.stem(word) for word in doc_filtered]
    
    # Task 5: Token Generation and Filtering
    doc_filtered_tokens = doc_stemmed
    
    # Create a dictionary to store the results
    transformations = {
        "Original Document (doc)": original_doc,
        "Text Normalization (doc_lower)": doc_lower,
        "Special Character Removal and Tokenization (doc_tokens)": doc_tokens,
#         "Stop Word Filtering (doc_filtered)": doc_filtered
        "Stop Word Filtering (doc_filtered)": doc_filtered,
        "Stemming (doc_stemmed)": doc_stemmed,
        "Token Generation and Filtering (doc_filtered_tokens)": doc_filtered_tokens
    }

    return transformations

def _text_transformation(original_doc):
    transformation = text_transformation(original_doc)
    return ' '.join(transformation["Token Generation and Filtering (doc_filtered_tokens)"])

# Example usage:
original_document = "I had a terrible experience with their customer service department."
text_transformation(original_document)
````

##### Domain-Specific Stop Words

The process of defining domain-specific stop words is crucial in text analysis to ensure that the resulting term-document matrix focuses on meaningful terms and reduces noise. In our analysis of consumer complaints, we've identified certain types of words that are not standard language-based stop words but should be treated as domain-specific stop words due to their limited value in our analysis. Here's an explanation of these domain-specific stop words:

1. **Company Names:** Consumer complaints often mention specific company names like Citibank, Wells Fargo, or Chase. While these names are important for understanding the context of the complaints, they are not valuable as terms for classification. Therefore, we categorize company names into domain-specific stop words to exclude them from our analysis.

2. **State Names:** State names, such as California, New York, or Texas, are also frequently mentioned in complaints. However, they typically do not contribute significantly to the classification of complaints into product categories. Therefore, we include state names in our stop words list to reduce their impact on our term-document matrix.

3. **Masked Personal Information:** The CFPB takes measures to protect personal and sensitive information in consumer complaints by masking them with placeholder characters like "xx," "xxx," or "xxxx." These masked characters appear consistently in many complaints but do not provide meaningful information for classification. Hence, we consider them as domain-specific stop words.

By including these domain-specific stop words in our list, we aim to refine the term-document matrix, enhance the accuracy of our classification models, and focus on the essential terms that truly impact the categorization of consumer complaints into product categories. This careful consideration of domain-specific stop words is a crucial step in optimizing our text analysis process.


````python
%%time
def remove_domain_specific_stop_words(complaints):
    """
    Remove domain-specific stop words like 'xxx', 'xx', 'xxxx', etc from a list of transformed complaints.

    Args:
        complaints (str): transformed complaints.

    Returns:
        str: A transformed complaints with domain-specific stop words removed.
    """
    # Initialize the maximum length to 0
    max_length = 0

    
    # Iterate through the words to find the maximum length
    for word in complaints.split():
        word_length = len(word)
        if word_length > max_length:
            max_length = word_length
            
    # Generate domain-specific stop words with variable lengths
    domain_specific_stop_words = ['x' * i for i in range(2, max_length + 1)]
    
    # Remove domain-specific stop words from the complaints list
    return ' '.join([word for word in complaints.split() if word not in domain_specific_stop_words])

balanced_df['Transformed_Complaints'] = balanced_df['Transformed_Complaints'].swifter.apply(remove_domain_specific_stop_words)

# Display the first few rows of the transformed dataset
balanced_df[['Consumer complaint narrative', 'Transformed_Complaints']].head()
````

#### 3.4.1.c Dimension Reduction

In the previous section, we transformed our consumer complaint narratives into a numerical format using the TF-IDF (Term Frequency-Inverse Document Frequency) technique. This transformation resulted in a high-dimensional term-document matrix where each row represents a complaint, and each column represents a unique term (word). While this representation captures the importance of terms within each document, it can be computationally expensive and may lead to overfitting in machine learning models.

To address this issue, we apply dimension reduction techniques to reduce the number of features (terms) while preserving the most important information. This step is essential for improving model efficiency and generalization.

##### 1. Principal Component Analysis (PCA)

One common approach for dimension reduction is Principal Component Analysis (PCA). PCA identifies the principal components (linear combinations of features) that explain the most variance in the data. By selecting a subset of these components, we can achieve dimensionality reduction. **After trying with PCA, still memory error issue persists.**

##### 2. Truncated SVD

Truncated Singular Value Decomposition ([Truncated SVD](https://analyticsindiamag.com/beginners-guide-to-truncated-svd-for-dimensionality-reduction/)) differs from Principal Component Analysis (PCA) in how it handles dimensionality reduction, making it more suitable for avoiding memory issues in certain scenarios.

1. **Sparse Data Handling:** Truncated SVD is particularly effective when dealing with sparse data, such as text data represented as TF-IDF matrices. Sparse data often contains many zero values, and Truncated SVD can efficiently work with such data, unlike PCA.

2. **Memory Efficiency:** PCA computes the covariance matrix of the data, which can be memory-intensive for high-dimensional datasets. Truncated SVD, on the other hand, directly operates on the data matrix, making it more memory-efficient. It doesn't require the expensive computation of the covariance matrix.

3. **Dimension Reduction:** Truncated SVD allows you to specify the number of components (k) to keep explicitly. This means you can control the level of dimensionality reduction, making it suitable for datasets with varying degrees of sparsity.

4. **Randomized Algorithms:** Truncated SVD often employs randomized algorithms, which provide approximate solutions with much lower memory requirements compared to exact methods. These algorithms are effective in reducing memory consumption while retaining data structure.

5. **Applications in Text Data:** Truncated SVD is commonly used in natural language processing tasks, where text data is inherently high-dimensional and sparse. It allows for efficient topic modeling and semantic analysis of text corpora.

In summary, Truncated SVD is a memory-efficient dimensionality reduction technique that excels in handling sparse data, making it a valuable alternative to PCA when memory issues are a concern, especially in text analysis and related domains.

`2.5. Deep Learning Model in TensorFlow`

**Hyperparameter Tuning using [Keras Tuner](https://keras.io/keras_tuner/)**: To perform hyperparameter tuning, we need to use `keras tuner` to define a search space and find the best model.

**Label Encoding target "Product" variable** is used to convert the textual categories in the target variable into numbers. This is necessary because machine learning models, including neural networks, require numerical values for both input features and output labels. Label encoding is a straightforward way to achieve this conversion, making the data compatible and efficient for model training and evaluation

````python
# !pip install keras-tuner

from keras_tuner import RandomSearch
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder

# Initialize label encoder
label_encoder = LabelEncoder()

# Fit and transform y_train and y_test
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

# Convert the data to TensorFlow tensors
X_train_tensor = tf.convert_to_tensor(X_train, dtype=tf.float32)
X_test_tensor = tf.convert_to_tensor(X_test, dtype=tf.float32)
y_train_tensor = tf.convert_to_tensor(y_train_encoded, dtype=tf.float32)
y_test_tensor = tf.convert_to_tensor(y_test_encoded, dtype=tf.float32)


def build_model(hp):
    """
    Build and compile a TensorFlow model for hyperparameter optimization.
    
    Args:
    hp (keras_tuner.HyperParameters): Hyperparameters to optimize.
    
    Returns:
    tf.keras.Model: Compiled model with specified hyperparameters.
    """
    
    # Initialize the model as a Sequential model
    model = tf.keras.Sequential()
    
    # Add the first Dense layer with hyperparameter-optimized number of units and 'relu' activation
    # The input shape is based on the number of features in the dataset
    model.add(tf.keras.layers.Dense(units=hp.Int('units_1', min_value=32, max_value=128, step=32),
                                    activation='relu', 
                                    input_shape=(X_train.shape[1],)))
    
    # Add the second Dense layer with hyperparameter-optimized number of units and 'relu' activation
    model.add(tf.keras.layers.Dense(units=hp.Int('units_2', min_value=16, max_value=64, step=16),
                                    activation='relu'))
    
    # Add the output Dense layer with softmax activation
    # The number of units is equal to the number of unique target labels
    model.add(tf.keras.layers.Dense(len(np.unique(y_train_encoded)), activation='softmax'))
    
    # Compile the model with a hyperparameter-optimized learning rate
    # The loss function is sparse_categorical_crossentropy and the metric is accuracy
    model.compile(optimizer=tf.keras.optimizers.Adam(hp.Choice('learning_rate', values=[1e-2, 1e-3, 1e-4])),
                  loss='sparse_categorical_crossentropy',
                  metrics=['accuracy'])
    
    return model

# Initialize the tuner
tuner = RandomSearch(build_model, objective='val_accuracy', max_trials=12, executions_per_trial=2)

# Perform hyperparameter search
tuner.search(X_train_tensor, y_train_tensor, epochs=10, validation_split=0.05)

# Get the optimal hyperparameters
best_hps = tuner.get_best_hyperparameters(num_trials=1)[0]
````

````python
def float_to_int(value):
    """
    Convert float values with trailing zeros to integers.
    
    Parameters:
    - value (float): The float value to be converted.
    
    Returns:
    - int: The integer representation of the input float.
    """
    return int(value)

def generate_performance_metrics(X_train, y_train, X_test, y_test, label_encoder, model_path=None):
    """
    Generate performance metrics including a performance table and an enhanced confusion matrix.
    
    Parameters:
    - X_train (array-like): The training feature matrix.
    - y_train (array-like): The training target vector.
    - X_test (array-like): The test feature matrix.
    - y_test (array-like): The test target vector.
    - label_encoder (LabelEncoder): The label encoder used for transforming target classes.
    - model_path (str, optional): The file path to the saved machine learning model.
    
    Returns:
    - performance_table (DataFrame): A table containing training and testing accuracies and classification errors.
    - df_cm (DataFrame): An enhanced confusion matrix with Class Precision and Class Recall.
    """
    
    # Load the saved model
    loaded_model = load(model_path)

    # Perform prediction on the training set
    train_predictions = loaded_model.predict(X_train)

    # Calculate the training accuracy
    train_accuracy = round(accuracy_score(y_train, train_predictions) * 100, 2)

    # Perform prediction on the test set
    test_predictions = loaded_model.predict(X_test)

    # Calculate the test accuracy
    test_accuracy = round(accuracy_score(y_test, test_predictions) * 100, 2)

    # Create the performance table
    performance_data = {
        'Criterion': ['Training Accuracy', 'Testing Accuracy', 
                      'Training Classification Error', 'Testing Classification Error'],
        'Value': [train_accuracy, test_accuracy, 100 - train_accuracy, 100 - test_accuracy]
    }
    performance_table = pd.DataFrame(performance_data)

    # Generate the confusion matrix
    cm = confusion_matrix(y_test, test_predictions)

    # Create a DataFrame for better visualization
    actual_labels = ['Actual: ' + str(label) for label in label_encoder.classes_]
    pred_labels = ['Pred: ' + str(label) for label in label_encoder.classes_]

    # Create DataFrame for the confusion matrix
    df_cm = pd.DataFrame(cm, index=actual_labels, columns=pred_labels)

    # Calculating Class Precision and Class Recall
    precision_row = (df_cm.values.diagonal() / df_cm.values.sum(axis=0)) * 100
    recall_col = (df_cm.values.diagonal() / df_cm.values.sum(axis=1)) * 100

    # Add Class Precision row and Class Recall column to the DataFrame
    df_cm.loc['Precision (in %)'] = precision_row
    df_cm['Recall (in %)'] = np.append(recall_col, np.nan)

    # Convert float values with trailing zeros to integers, excluding 'Recall (in %)'
    for col in df_cm.columns:
        if col not in ['Recall (in %)']:
            df_cm[col] = df_cm[col].apply(float_to_int)

    return performance_table, df_cm
````

#### 3.5.7 Observations on GLM Feature Importance

Let us implement a code which will first extract the `Singular Value Decomposition (SVD)` components and the feature names from the TF-IDF vectorizer, followed by obtaining the `Generalized Linear Model (GLM)` coefficients. Then, these coefficients will be mapped back to the original feature space through a dot product operation. A Pandas DataFrame will be created to store these mapped coefficients, indexed by class labels and columned by feature names. For each class label, the code will sort the coefficients in descending order to identify the top 30 most influential features, storing these in a dictionary. Finally, this dictionary will be converted into a DataFrame for easier interpretation and presentation.

````python
# Get SVD components and feature names
svd_components = svd.components_
feature_names = tfidf_vectorizer.get_feature_names_out()

# Get the coefficients from the GLM model
glm_coefficients = glm_model.coef_

# Map the GLM coefficients back to the original feature space
mapped_coefficients = np.dot(glm_coefficients, svd_components)

# Create a DataFrame for easier manipulation
df_mapped_coefficients = pd.DataFrame(mapped_coefficients, columns=feature_names, index=label_encoder.classes_)

# Initialize an empty dictionary to store sorted coefficients for each class
sorted_coefs_dict = {}

# For each class, sort the coefficients to get the most influential features
for class_label in label_encoder.classes_:
    sorted_coef = df_mapped_coefficients.loc[class_label].sort_values(ascending=False)
    sorted_coefs_dict[class_label] = list(sorted_coef.head(30).keys().values)
    
top_keywords = pd.DataFrame(sorted_coefs_dict)
top_keywords
````

## 7.2 Glossary

1. **GLM (Generalized Linear Models)**  
    - A flexible generalization of ordinary linear regression that allows for response variables that have error distribution models other than a normal distribution.

2. **TF-IDF (Term Frequency-Inverse Document Frequency)**  
    - A numerical statistic that reflects how important a word is to a document in a collection or corpus.

3. **SVD (Singular Value Decomposition)**  
    - A factorization of a real or complex matrix that generalizes the eigendecomposition of a square normal matrix.

4. **Naive Bayes**  
    - A classification technique based on Bayes' Theorem with an assumption of independence among predictors.

5. **Decision Trees**  
    - A decision support tool that uses a tree-like graph or model of decisions and their possible consequences.

6. **Random Forests**  
    - An ensemble learning method for classification, regression, and other tasks that operate by constructing a multitude of decision trees.

7. **Neural Networks**  
    - A series of algorithms that attempts to recognize underlying relationships in a set of data through a process that mimics the way the human brain operates.

8. **Precision and Recall**  
    - Precision is the number of True Positives divided by the number of True Positives and False Positives. Recall is the number of True Positives divided by the number of True Positives and the number of False Negatives.

9. **F1-Score**  
    - The harmonic mean of the precision and recall, where an F1 score reaches its best value at 1 (perfect precision and recall) and worst at 0.

10. **Label Encoder**  
    - A utility function to help normalize labels such that they contain only values between 0 and n_classes-1.

11. **Overfitting and Underfitting**  
    - Overfitting occurs when a statistical model or machine learning algorithm captures the noise of the data. Underfitting occurs when a statistical model or machine learning algorithm cannot capture the underlying trend of the data.

12. **Hyperparameter Tuning**  
    - The process of choosing a set of optimal hyperparameters for a learning algorithm.

13. **Cross-Validation**  
    - A resampling procedure used to evaluate machine learning models on a limited data sample.

14. **Confusion Matrix**  
    - A table used to describe the performance of a classification model on a set of data for which the true values are known.
